```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv(r"C:\Users\emmap\OneDrive\Documents\Datasets\cricket_matches.csv", header=1)
df.head(5)
```

```python
# rename columns
df = df.rename(columns= {'Mat':'Matches','NO':'Not_Outs', 'HS':'Highest_Inning_Score', 'BF':'Balls_Faced', 'SR': 'Batting_Strike_Rate'})
df.head()
```

```python
# Check for null values
df.isnull().sum()
```

```python
df[df['Balls_Faced'].isnull()==1]
```

```python
# replace null values
df['Balls_Faced'] = df['Balls_Faced'].fillna(0)
df.iloc[9]
```

```python
# find duplicates
df[df['Player'].duplicated()==1]
```

```python

df[df['Player'].isin(['GA Headley (WI)','GS Sobers (WI)', 'DR Jardine (ENG)'])]
```

```python
# drop duplicates
df = df.drop_duplicates(subset=['Player'])
df[df['Player'].isin(['GA Headley (WI)','GS Sobers (WI)', 'DR Jardine (ENG)'])]
```

```python
# split span into start and end dates
df['Span'].str.split('-', expand=True)
```

```python
# inserting split columns into dataframe
df['Rookie_Year'] = df['Span'].str.split('-', expand=True)[0]
df['Final_Year'] = df['Span'].str.split('-', expand=True)[1]
# drop Span
df.drop('Span', axis=1, inplace=True)
df.head()
```

```python
## Q: How to split up Country from Player
df['Player'].str.split('(', expand=True)
df['Country'] = df['Player'].str.split('(', expand=True)[1]
df.head()
```

```python
# split and remove the ) parenthesis from Country values
df['Country'] = df['Country'].str.split(')', expand=True)[0]
df.head()
```

```python
# split and remove country names from Player
df['Player'] = df['Player'].str.split('(', expand=True)[0]
df.head()
```

```python
# convert data types
df.dtypes
# convert Rookie Year, Final Year, Highest Inning Score, Balls Faced
df['Rookie_Year'] = df['Rookie_Year'].astype(int)
df['Final_Year'] = df['Final_Year'].astype(int)
```

```python
# create function for removing non-ints from column values
Orig_Balls_Faced = df['Balls_Faced']
Orig_Balls_Faced

def clean_numbers(column):
    column = column.str.replace(r'[^\d]', '', regex=True)
    column = column.fillna('')
    column = pd.to_numeric(column)
    return column

# apply function
df['Balls_Faced'] = clean_numbers(df['Balls_Faced'])
df['Highest_Inning_Score'] = clean_numbers(df['Highest_Inning_Score'])

df.dtypes
```

```python
# Q1: What is average career lengths of these players
df['Career_Length'] = df['Final_Year'] - df['Rookie_Year']
df.head()

# Answer
df['Career_Length'].mean()
```

```python
# Q2: What is average Batting Strike Rate for players who player over 10 years

# Answer
df[df['Career_Length'] > 10]['Batting_Strike_Rate'].mean()
```

```python
# Q3: What is number of players who played before 1960

# Answer
df[df['Final_Year'] < 1960]['Player'].count()
```

```python
# Q4: What is the max Highest Inning Scores by Country

# Answer
df.groupby('Country').agg({'Highest_Inning_Score' : 'max'}).sort_values(by='Highest_Inning_Score', ascending=False)
```

```python
# Q5: 100s, 50s, 0s average per Country

# Answer
df.groupby('Country')[['100', '50', '0',]].mean()
```
